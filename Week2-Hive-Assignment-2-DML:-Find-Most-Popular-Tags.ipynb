Let's count tag occurrence per year...

```python
%%writefile count_tags_per_year.hql

select count(*) as total, tag, year
    from stackoverflow_.posts
        lateral view explode(tags) lateral_table
        as tag
    group by tag, year
```

Let's answer question...

```python
%%writefile assignment_reply.hql

with resultats_2009 as (
    select count(*) as total_2009, tag as tag_2009
        from stackoverflow_.posts
            lateral view explode(tags) lateral_table
            as tag
        where post_type_id=1
        and year=2009
        group by tag, year
), resultats_2009_ranked as (
    select total_2009, tag_2009,
        rank() over (order by total_2009 desc) as rang_2009
    from resultats_2009
), resultats_2016 as (
    select count(*) as total_2016, tag as tag_2016
        from stackoverflow_.posts
            lateral view explode(tags) lateral_table
            as tag
        where post_type_id=1
        and year=2016
        group by tag, year
), resultats_2016_ranked as (
    select total_2016, tag_2016,
        rank() over (order by total_2016 desc) as rang_2016
    from resultats_2016
)
select tag_2009, rang_2016, rang_2009, total_2016, total_2009
from resultats_2009_ranked, resultats_2016_ranked
where tag_2009=tag_2016
order by rang_2016;
```

```python
!hive -f assignment_reply.hql
```

If I believe what I read on forums, we should prefer a solution with only one subquery and sum(if())...

```python
%%writefile assignment_reply_sumif.hql

USE stackoverflow_;

SELECT
    tag,
    rank() OVER (ORDER BY count2016 DESC) as rang_2016,
    rank() OVER (ORDER BY count2009 DESC) as rang_2009,
    total_2016,
    total_2009
FROM(
    SELECT
        tag,
        sum(if(year = '2009', 1, 0)) as total_2009,
        sum(if(year = '2016', 1, 0)) as total_2016
    FROM posts
    LATERAL VIEW explode(tags) tag as tag
    WHERE
        year in ('2016', '2009')
        AND post_type_id=1
    GROUP BY tag
) as tmp
ORDER BY rang_2016
LIMIT 10;
```
